#import libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/netflix_titles.csv")
```

```python
print("First 5 rows:")
display(df.head())
```

```python
print("\nDataset Info:")
print(df.info())
df.describe()
```

```python
print("\nShape of Dataset:", df.shape)
print("\nColumn name:",df.columns)
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())
```

# PART 2: Clean the Dataset


```python
df['director'] = df['director'].fillna("Not Available")
df['cast'] = df['cast'].fillna("Not Available")
df['country'] = df['country'].fillna("Not Available")
df['date_added'] = df['date_added'].fillna("Not Available")
df['rating'] = df['rating'].fillna("Not Available")
df['duration'] = df['duration'].fillna("Not Available")
print(df)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
print(df)
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
text_columns = df.select_dtypes(include='object').columns
for col in text_columns:
    df[col] = df[col].astype(str).str.lower().str.strip()
df
```

# PART 3: Univariate Analysis (Visuals + Stats)
# 1. Content Type


```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="type", order=df['type'].value_counts().index, palette="pastel")
plt.title("Distribution of Content Type")
plt.show()
```

```python
type_counts = df['type'].value_counts()
type_percent = (type_counts / len(df)) * 100
type_counts
type_percent
```

# 2. Content Rating

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, y="rating", order=df['rating'].value_counts().index, palette="muted")
plt.title("Distribution of Content Ratings")
plt.show()
```

```python
rating_counts = df['rating'].value_counts()
print(rating_counts)
```


# 3. Release Year

```python
plt.figure(figsize=(12,5))
sns.histplot(data=df, x="release_year", bins=30, kde=False, color="skyblue")
plt.title("Number of Releases per Year")
plt.show()
```

```python
release_counts = df['release_year'].value_counts().sort_index
print(release_counts)
```

# 4. Movie Duration


```python
movie_durations = (
    df[df['type'].str.lower() == 'movie']['duration']
    .str.replace(" min", "", case=False)
    .replace(["Not Available","not available"], pd.NA)  # turn 'Not Available' back into NaN
    .dropna()
    .astype(float)
)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=movie_durations, color="purple")
plt.title("Distribution of Movie Durations (in minutes)")
plt.xlabel("Duration (minutes)")
plt.show()
```

# 5. Top Countries


```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(12,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="cool")
plt.title("Top 10 Countries by Number of Netflix Titles")
plt.xlabel("Number of Titles")
plt.show()
```

# PART 4: Interpretations
# Q1: Content Type


```python
print(f"Most common type: {type_counts.idxmax().capitalize()}")
print(f"Movies: {type_percent['movie']:.2f}%, TV Shows: {type_percent['tv show']:.2f}%")
```

# Q2: Content Rating

```python
print(f"Most common rating: {rating_counts.idxmax().upper()}")
rare_ratings = rating_counts[rating_counts < 10].index.tolist()
print(f"Rare ratings: {rare_ratings}")
```

# Q3: Release Year

```python
print(f"Year with most releases: {release_counts().sort_index()}")
percent_after_2015 = (df['release_year'] > 2015).mean() * 100
print(f"Percentage released after 2015: {percent_after_2015:.2f}%")
```

# Q4: Duration


```python
print(f"Average movie length: {movie_durations.mean():.2f} min")
print(f"Shortest movie: {movie_durations.min()} min")
print(f"Longest movie: {movie_durations.max()} min")
```

# Q5: Country


```python
print(f"Top producing country: {top_countries.idxmax().capitalize()}")
low_rep_countries = df['country'].value_counts()[df['country'].value_counts() < 5].index.tolist()
print(f"Countries with very low representation: {low_rep_countries}")
```
